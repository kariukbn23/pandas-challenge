```python
import csv 
import pandas as pd 
```

```python
school_file = "../Resources/schools_complete.csv"

student_file = "../Resources/students_complete.csv"


df_school = pd.read_csv(school_file)

df_student = pd.read_csv(student_file)


school_data_complete = pd.merge(df_school, df_student, how="left", on=["school_name", "school_name"])


```

```python
total_schools = school_data_complete["School ID"].unique()

total_schools_final = len(total_schools)

total_schools_final
```

```python
total_students = school_data_complete["Student ID"].count()
total_students
```

```python
total_budget = school_data_complete["budget"].unique()
total_budget_final = sum(total_budget)
total_budget_final
```

```python
average_math_score = school_data_complete["math_score"].mean()

average_math_score = round(average_math_score, 6)

average_math_score
```

```python
average_reading_score = school_data_complete["reading_score"].mean()

average_reading_score = round(average_reading_score, 6)

average_reading_score
```

```python
passing_reading_scores_count = school_data_complete[school_data_complete["reading_score"] > 69.0].count()["reading_score"]

passing_reading_scores_count
```

```python
total_test_scores = school_data_complete["reading_score"].count()
total_test_scores
```

```python
passing_reading_scores_percentage = (passing_reading_scores_count/total_test_scores)*100
passing_reading_scores_percentage = round(passing_reading_scores_percentage, 6)
passing_reading_scores_percentage
```

```python
passing_math_scores_count = school_data_complete[school_data_complete["math_score"] > 69.0].count()["math_score"]

passing_math_scores_count
```

```python
passing_math_scores_percentage = (passing_math_scores_count/total_test_scores)*100
passing_math_scores_percentage = round(passing_math_scores_percentage, 6)
passing_math_scores_percentage
```

```python
overall_passing_rate_percentage = (passing_math_scores_percentage + passing_reading_scores_percentage)/2

overall_passing_rate_percentage = round(overall_passing_rate_percentage, 6)

overall_passing_rate_percentage
```

```python
district_summary_draft1 = school_data_complete[["School ID", "Student ID", "budget", "reading_score", "math_score"]]
# district_summary_draft1
```

```python
district_summary_draft2 = district_summary_draft1.rename(columns={"School ID" : "Total Schools", "Student ID" : "Total Students", "budget" : "Total Budget", "math_score" : "Average Math Score", "reading_score" : "Average Reading Score"})
# district_summary_draft2
```

```python
district_summary_final = pd.DataFrame({"Total Schools": [total_schools_final], 
                                        "Total Students": [total_students], 
                                        "Total Budget": [total_budget_final], 
                                        "Average Math Score": [average_math_score], 
                                        "Average Reading Score": [average_reading_score], 
                                        "% Passing Math": [passing_math_scores_percentage], 
                                        "% Passing Reading": [passing_reading_scores_percentage], 
                                        "% Overall Passing Rate": [overall_passing_rate_percentage]})
district_summary_final
```

```python
school_data_complete_new = pd.merge(df_school, df_student, how="left", on=["school_name", "school_name"])

school_data_complete_new.head()
```

```python
school_data_complete_new["Per Student Budget"] = school_data_complete_new["budget"] / school_data_complete_new["size"]

school_data_complete_newA = school_data_complete_new.groupby("school_name").mean()

school_data_complete_newA
```

```python
school_data_complete_newA["Avg. Reading Score"] = school_data_complete_newA.groupby("school_name")["reading_score"].mean()

school_data_complete_newA
```

```python
school_data_complete_newA["Avg. Math Score"] = school_data_complete_newA.groupby("school_name")["math_score"].mean()

school_data_complete_newA
```

```python
school_data_complete_newB = school_data_complete_newA[["size", "budget", "Per Student Budget", "Avg. Math Score", "Avg. Reading Score"]]

school_data_complete_newB
```

```python
school_data_complete_newC = school_data_complete_newB.rename(columns={"school_name":"School Name", 
                                                                  "type":"School Type",
                                                                  "size": "Total Students", 
                                                                  "budget": "Total Budget"})

school_data_complete_newC
```

```python
per_school_passing_math_scores = school_data_complete_new[(school_data_complete_new["math_score"] >= 70)].groupby(["school_name"]).count()["student_name"]/school_data_complete_new.groupby(["school_name"]).count()["student_name"]*100
per_school_passing_math_scores


```

```python
per_school_passing_reading_scores = school_data_complete_new[(school_data_complete_new["reading_score"] >= 70)].groupby(["school_name"]).count()["student_name"]/school_data_complete_new.groupby(["school_name"]).count()["student_name"]*100
per_school_passing_reading_scores
```

```python
school_data_complete_newC["% Passing Math"] = per_school_passing_math_scores

school_data_complete_newD = school_data_complete_newC.groupby("school_name").mean()

school_data_complete_newD
```

```python
school_data_complete_newD["% Passing Reading"] = per_school_passing_reading_scores

school_data_complete_newE = school_data_complete_newD.groupby("school_name").mean()

school_data_complete_newE
```

```python
school_data_complete_newE["% Overall Passing"] = (per_school_passing_reading_scores+per_school_passing_math_scores)/2

school_data_complete_newF = school_data_complete_newE.groupby("school_name").mean()

school_data_complete_newF
```

```python
# sort_by_life = gapminder.sort_values('lifeExp',ascending=False)

school_data_Top5_OverallPassingRate_final = school_data_complete_newF.sort_values("% Overall Passing", ascending=False)
school_data_Top5_OverallPassingRate_final.head()
```

```python
school_data_Bottom5_OverallPassingRate_final = school_data_complete_newF.sort_values("% Overall Passing")
school_data_Bottom5_OverallPassingRate_final.head()
```

```python
school_data_complete_new4["% Overall Passing"] = (charter_district_passing_math_percentages+charter_district_passing_reading_percentages)/2

school_data_complete_new5 = school_data_complete_new4.groupby("type").mean()

school_data_complete_new5
```

```python
charter_district_table_draft1 = school_data_complete_new5[["math_score", "reading_score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

charter_district_table_draft1
```
