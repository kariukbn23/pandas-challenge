```python
#importing dependencies 
import csv 
import pandas as pd 
```

```python
# creating variables to store csv paths and files and then creating a data frame that mergers my multiple csvs
school_file = "../Resources/schools_complete.csv"

student_file = "../Resources/students_complete.csv"


df_school = pd.read_csv(school_file)

df_student = pd.read_csv(student_file)


school_data_complete = pd.merge(df_school, df_student, how="left", on=["school_name", "school_name"])


```

```python
#Variable to check the unique valeus in the school ID column of data set and then performing len funciton to determine the number of items in that array
total_schools = school_data_complete["School ID"].unique()

total_schools_final = len(total_schools)

total_schools_final
```

```python
# same logic as before except performed a .count to determine the total number of students in this column given that there was a unique student id for each data point

total_students = school_data_complete["Student ID"].count()
total_students
```

```python
# same logic as total schools except since I needed a total of the unique objects in the array I performed the sum operation

total_budget = school_data_complete["budget"].unique()
total_budget_final = sum(total_budget)
total_budget_final
```

```python
#To calculate average math score of the distrtic I just did a .mean on the column value of focus and then rounded to ensure formatting was the same as exemplar
average_math_score = school_data_complete["math_score"].mean()

average_math_score = round(average_math_score, 6)

average_math_score
```

```python
# same logic as previous input
average_reading_score = school_data_complete["reading_score"].mean()

average_reading_score = round(average_reading_score, 6)

average_reading_score
```

```python
#In order to determine the number of score that were passing I had to perform a .count that filter through the df[df["specific object"]] frame work that allows you filter within a filter for specific criteria within a set of column values
passing_reading_scores_count = school_data_complete[school_data_complete["reading_score"] > 69.0].count()["reading_score"]

passing_reading_scores_count
```

```python
# same logic as total_students
total_test_scores = school_data_complete["reading_score"].count()
total_test_scores
```

```python
# calculations for determining the percentage of reading scores that were passing, relative to the total scores provided
passing_reading_scores_percentage = (passing_reading_scores_count/total_test_scores)*100
passing_reading_scores_percentage = round(passing_reading_scores_percentage, 6)
passing_reading_scores_percentage
```

```python
# For the next 3 inputs the same logic as used to find the reading evaluations above
passing_math_scores_count = school_data_complete[school_data_complete["math_score"] > 69.0].count()["math_score"]

passing_math_scores_count
```

```python
passing_math_scores_percentage = (passing_math_scores_count/total_test_scores)*100
passing_math_scores_percentage = round(passing_math_scores_percentage, 6)
passing_math_scores_percentage
```

```python
overall_passing_rate_percentage = (passing_math_scores_percentage + passing_reading_scores_percentage)/2

overall_passing_rate_percentage = round(overall_passing_rate_percentage, 6)

overall_passing_rate_percentage
```

```python
#Created a new data frame to pass only the specific column values that I wanted to display
district_summary_draft1 = school_data_complete[["School ID", "Student ID", "budget", "reading_score", "math_score"]]
district_summary_draft1.head()
```

```python
#Renamed column headers using method learning in class 
district_summary_draft2 = district_summary_draft1.rename(columns={"School ID" : "Total Schools", "Student ID" : "Total Students", "budget" : "Total Budget", "math_score" : "Average Math Score", "reading_score" : "Average Reading Score"})
district_summary_draft2.head()
```

```python
#Lastly created a final data frame that included a dictionary of keys that stored values for the all the variables found in the previous parts of my code
district_summary_final = pd.DataFrame({"Total Schools": [total_schools_final], 
                                        "Total Students": [total_students], 
                                        "Total Budget": [total_budget_final], 
                                        "Average Math Score": [average_math_score], 
                                        "Average Reading Score": [average_reading_score], 
                                        "% Passing Math": [passing_math_scores_percentage], 
                                        "% Passing Reading": [passing_reading_scores_percentage], 
                                        "% Overall Passing Rate": [overall_passing_rate_percentage]})
district_summary_final
```

```python
#For my second task I created a new data frame using the same merging techniques in the opening part of this hw
school_data_complete_new = pd.merge(df_school, df_student, how="left", on=["school_name", "school_name"])

school_data_complete_new.head()
```

```python
#Created a new column value within this data set that required dividing the values associated in two separate columns
school_data_complete_new["Per Student Budget"] = school_data_complete_new["budget"] / school_data_complete_new["size"]


# Performed my first groupby function to organize data based on the unique values found in the school name column. The .mean automatically generated the average of each of the values assoicated with each school name
school_data_complete_newA = school_data_complete_new.groupby("school_name").mean()

school_data_complete_newA
```

```python
#Even though I did not need to create a new column value for avg. reading scores, I did just to ensure the information calculated in the reading_score column was accurate
school_data_complete_newA["Avg. Reading Score"] = school_data_complete_newA.groupby("school_name")["reading_score"].mean()

school_data_complete_newA
```

```python
# Same logic as above
school_data_complete_newA["Avg. Math Score"] = school_data_complete_newA.groupby("school_name")["math_score"].mean()

school_data_complete_newA
```

```python
#Created a new data frame that called only the column values I needed to be passed
school_data_complete_newB = school_data_complete_newA[["size", "budget", "Per Student Budget", "Avg. Math Score", "Avg. Reading Score"]]

school_data_complete_newB
```

```python
#Performed the necessary renaming to match the exemplars provided
school_data_complete_newC = school_data_complete_newB.rename(columns={"school_name":"School Name", 
                                                                  "type":"School Type",
                                                                  "size": "Total Students", 
                                                                  "budget": "Total Budget"})

school_data_complete_newC
```

```python
#To determine the percentage of scores within a specific school that were passing I needed to do a double filter as I did in a prior task by doing the df(df["specific object"]) approach. In addition becuase there was a specific critiera I had to include that in closed brackets. What was new about this task was that I needed to also group by the specific schools that it provides me with one overall mean of the evaluation of the scores provided. The technique of .groupby("original column")["column that I wanted the mean of"].mean is used multiple times throught my codes
per_school_passing_math_scores = school_data_complete_new[(school_data_complete_new["math_score"] >= 70)].groupby(["school_name"]).count()["student_name"]/school_data_complete_new.groupby(["school_name"]).count()["student_name"]*100
per_school_passing_math_scores


```

```python
# Same logic as above. Additionally the reason why I stored this informaiton as a variable is because later on when I create a new data frame, I will be able to pass these variables as column values 
per_school_passing_reading_scores = school_data_complete_new[(school_data_complete_new["reading_score"] >= 70)].groupby(["school_name"]).count()["student_name"]/school_data_complete_new.groupby(["school_name"]).count()["student_name"]*100
per_school_passing_reading_scores
```

```python
#Refer to logic in previous input. The explicit process here is that I created a new column of my data frame using the variable created above. Then I created an additonal data frame that was groupby by my designated filter. 
school_data_complete_newC["% Passing Math"] = per_school_passing_math_scores

school_data_complete_newD = school_data_complete_newC.groupby("school_name").mean()

school_data_complete_newD
```

```python
#Same logic as above
school_data_complete_newD["% Passing Reading"] = per_school_passing_reading_scores

school_data_complete_newE = school_data_complete_newD.groupby("school_name").mean()

school_data_complete_newE
```

```python
#Same logic as preivous two inputs, except I had to take the average of the two column values that I created. This was similar to when I did the Per Student Budget in the first task
school_data_complete_newE["% Overall Passing"] = (per_school_passing_reading_scores+per_school_passing_math_scores)/2

school_data_complete_newF = school_data_complete_newE.groupby("school_name").mean()

school_data_complete_newF
```

```python
# Final table for sort of Top 5 Overall passing rates by school can be found below. The (, ascending=False) is something I researched to apply the filter accross my designated data frame

school_data_Top5_OverallPassingRate_final = school_data_complete_newF.sort_values("% Overall Passing", ascending=False)
school_data_Top5_OverallPassingRate_final.head()
```

```python
#Same logic as above. When the (ascending = False) is taken away the .sort_values automatically sorts the values from least to greatest or another way to think about it is it would do a ascending = True
school_data_Bottom5_OverallPassingRate_final = school_data_complete_newF.sort_values("% Overall Passing")
school_data_Bottom5_OverallPassingRate_final.head()
```
